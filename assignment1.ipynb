```python
import myCodes.crank as crank
import numpy as np

def g(x):
    return 4*x - (x**2)/2

L = 8  # Length of the rod
T = 1  # Total time
Nx = 100  # Number of spatial points
Nt = 1000  # Number of time steps
alpha = (T/Nt)/((L/Nx)**2)
print("Alpha = ", alpha)
#alpha is determined by solving the problem manually keeping in mind the stability of the equation
crank.main(L,T,Nx,Nt,g)
#It takes around 25-30 sec to display the output. Please wait....
###Heat is maximum at the center and dissipates as we move forward in time
```

```python
import numpy as np
import myCodes.poisson as poi
# Define parameters
nx, ny = 6, 6
Lx, Ly = 2, 1

# Create grid
x = np.linspace(0, Lx, nx)
y = np.linspace(0, Ly, ny)

# Initialize solution matrix
u = np.zeros((nx, ny))

# Set boundary conditions
u[:, 0] = x
u[:, -1] = x * np.exp(1)
u[0, :] = 0
u[-1, :] = 2 * np.exp(y)

poi.main(Lx,Ly,nx,ny,u)

```
